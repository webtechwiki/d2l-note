# 线性回归

线性回归是机器学习最基础的一个模型，也是我们理解所有深度学习模型的基础，所以我们先从线性回归开始。

## 1. 什么是线性回归

我们先看个例子：

在美国，有一份关于房价的数据集，其中包含了美国一些城市的房价和其他信息。我们现在想买房，但是我们不知道房价是多少，我们可以通过这份数据集来预测一下房价。我们可以使用这个数据集来训练一个模型，输入城市的信息，就可以预测房价。

我们把这个问题抽象成一个机器学习问题，我们把城市的信息叫做x，我们把房价叫做y，我们的目标就是通过训练数据集，得到一个函数，输入x，就可以得到y。这就是一个最简单的机器学习问题，也就是最简单的线性回归问题。

## 2. 一个简化模型

假设1：我们假设影响房价的关键因素是卧室的数量、卫生间的数量、房子的面积，记为x<sub>1</sub>，x<sub>2</sub>，x<sub>3</sub>。

假设2：成交价是关键因素的加权和 y = w<sub>1</sub>x<sub>1</sub> + w<sub>2</sub>x<sub>2</sub> + w<sub>3</sub>x<sub>3</sub> + b。权重和偏差的实际值在后面再决定。

## 3. 线性模型

我们把2中的问题拓展到一般化的线性模型，

给定n维输入 x = [ x<sub>1</sub>，x<sub>2</sub>，...，x<sub>n</sub> ]<sup>T</sup>，b

线性模型有一个n维权重和一个标量偏差 w = [ w<sub>1</sub>，w<sub>2</sub>，...，w<sub>n</sub> ]，b

输出是输入的加权和 y = w<sub>1</sub>x<sub>1</sub> + w<sub>2</sub>x<sub>2</sub> + ... + w<sub>n</sub>x<sub>n</sub> + b

向量版本：y = < w，x > + b


线性模型可以看做单层的神经网络，输入层的神经元个数是n，输出层的神经元个数是1。

## 6. 总结

- 线性回归是对n维输入的加权，外加一个偏差
- 使用平方损失来衡量预测值和真实值的差异
- 线性回归有显示解
- 线性内回归可以看做单层神经网络
